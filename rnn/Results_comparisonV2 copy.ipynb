```python
# path

#path of the tested models 
def models_path(j) : return f"tests/models_test/epochs={nbr_epoch}/models_units={units(j)}_batch={nbr_batch}_lr={lr}.h5"

#path of the qconverted models
def qmodels_path1(j, i): return f"tests/qmodels_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch1}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.h5"
def qpredicts_path1(j,i): return f"tests/qpredicts_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch1}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.npy"

def qmodels_path2(j, i): return f"tests/qmodels_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch2}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.h5"
def qpredicts_path2(j,i): return f"tests/qpredicts_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch2}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.npy"

def qmodels_path3(j, i): return f"tests/qmodels_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch3}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.h5"
def qpredicts_path3(j,i): return f"tests/qpredicts_test/qconverted/qmodels_units={units(j)}_epoch={nbr_conv_epoch3}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.npy"

#path of the qtrained models
def qtrained_models_path(j, i): return f"tests/qmodels_test/qtrained/qmodels_units={units(j)}_epoch={nbr_epoch}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.h5"
def qtrained_predicts_path(j, i):  return f"tests/qpredicts_test/qtrained/qmodels_units={units(j)}_epoch={nbr_epoch}_batch={nbr_batch}_lr={lr},<{bit_lenghts(i)},{integer}>.npy"

#path of the qref models 
def qrefmodels_path (i) : return f"qmodels_epoch={nbr_conv_epoch2}/qmodels<{bit_lenghts(i)},{integer}>.h5"
def qrefpredicts_path (i) : return f"qpredicts_epoch={nbr_conv_epoch2}/qpredicts<{bit_lenghts(i)},{integer}>.npy"
```

```python

for j in range (units_range):
    for i in range(bits_range):

        qmodel1 = qkeras.utils.load_qmodel(qmodels_path1(j,i))        
        exist(qpredicts_path1(j,i), qmodel1)
        
        qmodel2 = qkeras.utils.load_qmodel(qmodels_path2(j,i))        
        exist(qpredicts_path2(j,i), qmodel2)        
        
        qmodel3 = qkeras.utils.load_qmodel(qmodels_path3(j,i))        
        exist(qpredicts_path3(j,i), qmodel3)
```

```python
unit_nbr = []
for j in range (units_range):
    unit_nbr.append(f'{units(j)}')

    loss1 = [] 
    loss2 = []
    loss3 = []
    loss4 = []  
    bit_nbr = []
    for i in range(bits_range):
        bit_nbr.append(bit_lenghts(i))
        y_qkeras1 = np.load(qpredicts_path1(j,i))
        y_qkeras2 = np.load(qpredicts_path2(j,i))
        y_qkeras3= np.load(qpredicts_path3(j,i))
        
        loss1.append(mse(y_test,y_qkeras1))        
        loss2.append(mse(y_test,y_qkeras2))
        loss3.append(mse(y_test,y_qkeras3))
    
    plt.figure()
    plt.plot(loss1)
    plt.plot(loss2)
    plt.plot(loss3)
    plt.legend(['4 epochs ','8 epochs','12 epochs'])
    plt.xticks(np.arange(len(bit_nbr)), [f"<{Bit},{integer}>" for Bit in bit_nbr])
    plt.xlabel('units settings')
    plt.yscale('log')
    plt.ylabel('MSE')
    plt.title(f'MSE qkeras with {units(j)} units, comparison with differente bitwidth \n network used batchsize = {nbr_batch}, lr={lr}')
plt.show()
```
